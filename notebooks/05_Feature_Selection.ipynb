1. Carga e inspección inicial del dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

DATA_PATH = Path("../data/processed/passwords_dataset_target_externo.csv")


# Configuración visual
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid", palette="muted")

#  Carga del dataset
df = pd.read_csv(DATA_PATH)

# Inspección general
print("Dimensiones del dataset:", df.shape)
print("\n Vista general de las primeras filas:")
display(df.head())

# Tipos de datos y valores nulos
print("\n Información general del dataset:")
df.info()

print("\n Conteo de valores nulos por columna:")
display(df.isnull().sum().sort_values(ascending=False))

# Estadísticas descriptivas (solo para variables numéricas)
print("\n Estadísticas descriptivas:")
display(df.describe().T)

# Distribución de la variable objetivo
if 'target_bin_zxcvbn' in df.columns:
    plt.figure(figsize=(4, 4))
    sns.countplot(x='target_bin_zxcvbn', data=df, palette="coolwarm")
    plt.title("Distribución de la variable objetivo (target_bin_zxcvbn)")
    plt.xlabel("Clase")
    plt.ylabel("Cantidad de contraseñas")
    plt.show()
    
    ratio = df['target_bin_zxcvbn'].value_counts(normalize=True)
    print("\n Proporción de clases:")
    print(ratio)
else:
    print(" No se encontró la columna 'target_bin_zxcvbn'. Verifica el dataset.")

# Detección rápida de duplicados
duplicates = df.duplicated().sum()
print(f"\n Registros duplicados detectados: {duplicates}")
```

2. Análisis de correlación

```python
numeric_df = df.select_dtypes(include=[np.number]).copy()
print("Número de variables numéricas:", numeric_df.shape[1])

# Calcular la matriz de correlación (Spearman es más robusta ante no linealidades)
corr_matrix = numeric_df.corr(method='spearman')

# Visualizar la matriz de correlación
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    cbar_kws={'label': 'Coeficiente de correlación (Spearman)'}
)
plt.title(" Matriz de correlación entre variables numéricas", fontsize=14)
plt.show()

# Identificar pares de variables con alta correlación
# Se considera redundancia cuando |ρ| > 0.85
threshold = 0.85
corr_pairs = (
    corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    .stack()
    .reset_index()
    .rename(columns={'level_0': 'Feature_1', 'level_1': 'Feature_2', 0: 'Correlation'})
)

high_corr = corr_pairs[abs(corr_pairs['Correlation']) > threshold]
print(f"\n Variables altamente correlacionadas (|ρ| > {threshold}):")
display(high_corr.sort_values(by='Correlation', ascending=False))

# Crear lista de features a eliminar
to_drop = set()
for col in high_corr['Feature_2']:
    to_drop.add(col)

print(f"\n Variables sugeridas para eliminar por alta correlación ({len(to_drop)} en total):")
print(to_drop)

#Crear dataframe reducido sin las variables redundantes
df_reduced_corr = numeric_df.drop(columns=list(to_drop), errors='ignore')

print("\n Dimensiones después de eliminar variables correlacionadas:", df_reduced_corr.shape)
```

3. Análisis de varianza

```python
# Filtrado por varianza
from sklearn.feature_selection import VarianceThreshold

# 0) Elegir la fuente base: si df_reduced_corr existe úsalo, si no usa df
source_df = df_reduced_corr if 'df_reduced_corr' in globals() else df.copy()

# 1) Separar X e y de forma segura
if 'target_bin_zxcvbn' in source_df.columns:
    y = source_df['target_bin_zxcvbn'].astype(int)
    X_base = source_df.drop(columns=['target_bin_zxcvbn'])
else:
    assert 'target_bin_zxcvbn' in df.columns, "No se encuentra 'target_bin_zxcvbn' en los dataframes."
    y = df['target_bin_zxcvbn'].astype(int)
    X_base = source_df  # continuar con las columnas disponibles

print("Dimensiones base de X:", X_base.shape)

# 2) Trabajar solo con columnas numéricas
X_num = X_base.select_dtypes(include=[np.number]).copy()
print("Columnas numéricas en X:", X_num.shape[1])

# 3) Varianza individual
variances = X_num.var().sort_values(ascending=True)
print("\nVarianza por feature (asc):")
display(variances)

# 4) Filtro de baja varianza (ajusta el umbral si lo ves conveniente)
threshold = 0.001  # para binarios, elimina los que están hiper desbalanceados
selector = VarianceThreshold(threshold=threshold)
selector.fit(X_num)

low_variance_features = X_num.columns[~selector.get_support()]
print(f"\nFeatures con varianza < {threshold} ({len(low_variance_features)}):")
print(list(low_variance_features))

# 5) Conservar solo las columnas con varianza suficiente
X_reduced_var = X_num.loc[:, selector.get_support()]
print("\nDimensiones después del filtro de varianza:", X_reduced_var.shape)

# 6) Reconstruir dataset final (X_reduced_var + y)
df_reduced_var = pd.concat([X_reduced_var, y.reset_index(drop=True)], axis=1)
print("Dimensiones finales (con target):", df_reduced_var.shape)

```

```python
#Selección univariante (MI y Chi²)
from sklearn.feature_selection import mutual_info_classif, chi2
from sklearn.preprocessing import MinMaxScaler

# Usa el dataframe 
df_base = df_reduced_var

X = df_base.drop(columns=['target_bin_zxcvbn'])
y = df_base['target_bin_zxcvbn'].astype(int)

# 1) Mutual Information (acepta valores negativos/continuos sin escalar)
mi_scores = mutual_info_classif(X.fillna(X.median(numeric_only=True)), y, random_state=42)
mi = pd.Series(mi_scores, index=X.columns, name='MI').sort_values(ascending=False)

# 2) Chi² (requiere no-negativos y es sensible a escala -> MinMax)
X_pos = X.fillna(X.median(numeric_only=True))
scaler = MinMaxScaler()
X_chi = pd.DataFrame(scaler.fit_transform(X_pos), columns=X_pos.columns)
chi_scores, _ = chi2(X_chi, y)
chi = pd.Series(chi_scores, index=X.columns, name='Chi2').sort_values(ascending=False)

# 3) Tabla combinada y ranking promedio
rank_df = pd.concat([mi, chi], axis=1)
rank_df['rank_MI'] = rank_df['MI'].rank(ascending=False, method='dense')
rank_df['rank_Chi2'] = rank_df['Chi2'].rank(ascending=False, method='dense')
rank_df['rank_avg'] = rank_df[['rank_MI','rank_Chi2']].mean(axis=1)
rank_df = rank_df.sort_values('rank_avg')

print("Top 20 por ranking combinado (MI + Chi²):")
display(rank_df.head(20))

# 4) Selección top-k (ajusta k)
k = 20
topk = rank_df.head(k).index.tolist()

# Asegura que las protegidas se incluyan (si no entraron por top-k)
protected = {
    # Morfología
    'length', 'num_upper', 'num_lower', 'num_digits', 'num_symbols', 'charset_size',
    # Entropía
    'entropy_combinatorial',
    # Patrones
    '_pattern_count', 'has_seq_num', 'has_keyboard', '_has_long_repeats',
    'has_repeat', '_max_same_char_run',
    # Contexto semántico
    '_has_year4', '_has_date8', 'has_common', '_is_email_strict'
}
for p in protected:
    if p in X.columns and p not in topk:
        topk.append(p)

print(f"\nFeatures seleccionadas (top-{k} + protegidas si aplica):")
print(topk)

# 5) Dataset final de features seleccionadas para modelado
df_features_selected = pd.concat([X[topk], y], axis=1)
print("Dimensiones df_features_selected:", df_features_selected.shape)


```

```python
from pathlib import Path

# Ruta del archivo de salida
out_path = Path("../data/processed/passwords_features_selected_zxcvbn.csv")

# Crear la carpeta si no existe
out_path.parent.mkdir(parents=True, exist_ok=True)

# Guardar el dataset
df_features_selected.to_csv(out_path, index=False, encoding="utf-8")
print(f" Guardado correctamente en: {out_path.resolve()}")
```
