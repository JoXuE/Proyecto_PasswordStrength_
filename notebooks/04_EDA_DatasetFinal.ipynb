1. Estadistica descriptiva

```python
import pandas as pd

# --- 1. Cargar dataset ---
DATA_PATH = "../data/processed/passwords_dataset_target_externo.csv"
df = pd.read_csv(DATA_PATH)

# --- 2. Dimensiones ---
print("Dimensiones del dataset:", df.shape)

# --- 3. Nombres de columnas y tipos ---
print("\nTipos de datos:")
print(df.dtypes)

# --- 4. Primeras filas ---
print("\nPrimeras filas del dataset:")
display(df.head())

# --- 5. Valores faltantes ---
print("\nValores nulos por columna:")
print(df.isna().sum())

# --- 6. Resumen estadístico general ---
print("\nResumen estadístico:")
display(df.describe(include="all"))
```

2.Analisis Descrptivo

```python
import matplotlib.pyplot as plt
import seaborn as sns


#2.1. Distribución de la longitud de contraseñas
plt.figure(figsize=(10,6))
sns.histplot(df["length"], bins=30, kde=True)
plt.title("Distribución de la longitud de contraseñas")
plt.xlabel("Longitud")
plt.ylabel("Frecuencia")
plt.show()

print(df["length"].describe())
```

```python
#2.2. Composición de caracteres en contraseñas

composition = {
    "Solo números": (df["num_lower"]==0) & (df["num_upper"]==0) & (df["num_symbols"]==0),
    "Solo letras": (df["num_digits"]==0) & (df["num_symbols"]==0),
    "Contiene mayúsculas": (df["num_upper"]>0),
    "Contiene dígitos": (df["num_digits"]>0),
    "Contiene símbolos": (df["num_symbols"]>0)
}

comp_counts = {k: v.mean() for k,v in composition.items()}

# Pasar a DataFrame
comp_df = pd.DataFrame.from_dict(comp_counts, orient="index", columns=["Proporción"])
print(comp_df)

# Visualización
plt.figure(figsize=(8,5))
sns.barplot(x=comp_df.index, y=comp_df["Proporción"])
plt.title("Composición de caracteres en contraseñas")
plt.ylabel("Proporción")
plt.xlabel("tipos de composición de caracteres.")
plt.xticks(rotation=30)
plt.show()
```

```python
#2.3 Resumen de composición básica

plt.figure(figsize=(12,6))

# --- Subgráfico 1: target_bin = 0 ---
plt.subplot(1,2,1)
sns.boxplot(data=df[df["target_bin_zxcvbn"]==0][["num_digits","num_upper","num_lower","num_symbols"]])
plt.title("Distribución caracteres - Contraseñas débiles (target_bin=0)")
plt.ylabel("Cantidad")

# --- Subgráfico 2: target_bin = 1 ---
plt.subplot(1,2,2)
sns.boxplot(data=df[df["target_bin_zxcvbn"]==1][["num_digits","num_upper","num_lower","num_symbols"]])
plt.title("Distribución caracteres - Contraseñas fuertes (target_bin=1)")
plt.ylabel("Cantidad")

plt.tight_layout()
plt.show()

```

```python
# Seleccionar solo columnas de interés
cols = ["num_digits","num_upper","num_lower","num_symbols","target_bin_zxcvbn"]

# Agrupar por target_bin y calcular estadísticas
summary = df[cols].groupby("target_bin_zxcvbn").agg(["mean","median"]).round(2)

# Renombrar índices para claridad
summary.index = summary.index.map({0:"Debil (0)", 1:"Fuerte (1)"})

summary
```

3. Patrones triviales

```python
patterns = {
    "Secuencia letras": df["has_seq_alpha"].mean(),
    "Secuencia numérica": df["has_seq_num"].mean(),
    "Repeticiones": df["has_repeat"].mean(),
    "Patrón teclado": df["has_keyboard"].mean(),
    "Año/fecha": df["has_year"].mean(),
    "Palabra común": df["has_common"].mean()
}

patterns_df = pd.DataFrame.from_dict(patterns, orient="index", columns=["Proporción"])
print(patterns_df)

plt.figure(figsize=(8,5))
sns.barplot(x=patterns_df.index, y=patterns_df["Proporción"])
plt.title("Patrones triviales en contraseñas")
plt.ylabel("Proporción")
plt.xlabel("Tipos de patrones triviales")
plt.xticks(rotation=30)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#columnas de patrones triviales 
trivial_cols = ["has_seq_alpha","has_seq_num","has_repeat",
                "has_keyboard","has_year","has_common"]

#indicador: ¿tiene ≥1 patrón trivial? 
df["trivial_any"] = (df[trivial_cols].sum(axis=1) >= 1)

# --- tabla de conteos por grupo (target_bin x trivial_any) ---
ct = df.pivot_table(index="target_bin_zxcvbn", columns="trivial_any",
                    values="pattern_template", aggfunc="count").fillna(0)

# aseguramos columnas False/True en orden
ct = ct.reindex(columns=[False, True]).astype(int)
ct.columns = ["Ninguno", "≥1 patrón"]

# --- porcentajes por fila (dentro de cada target) ---
pct = (ct.div(ct.sum(axis=1), axis=0) * 100).round(2)

print("Conteos:")
display(ct)
print("\nPorcentajes (%):")
display(pct)

# métrica extra útil: nº medio de patrones triviales por grupo
mean_trivial_per_group = df.groupby("target_bin_zxcvbn")[trivial_cols].mean().sum(axis=1).round(3)
print("\nPromedio de patrones triviales por contraseña (por target):")
print(mean_trivial_per_group.rename({0:"Débil (0)", 1:"Fuerte (1)"}))
```

```python
# etiquetas (0 = débil, 1 = fuerte)
row_labels = ["Débil (0)", "Fuerte (1)"]
plot_df = pct.copy()
plot_df.index = row_labels

ax = plot_df.plot(kind="bar", stacked=True, figsize=(8,5))
ax.set_title("Patrones triviales por grupo (porcentaje)")
ax.set_ylabel("Porcentaje (%)")
ax.set_xlabel("Grupo")
ax.legend(title="Patrones", loc="upper right")
for p in ax.patches:
# etiqueta dentro de la barra si el valor es significativo
    width, height = p.get_width(), p.get_height()
    if height > 5:
        ax.text(p.get_x()+width/2, p.get_y()+height/2, f"{height:.0f}%",
                ha="center", va="center", fontsize=9)
plt.tight_layout()
plt.show()
```

```python
#filtrar solo contraseñas fuertes 
df_weak = df[df["target_bin_zxcvbn"] == 1].copy()

#contar cuántos patrones triviales cumple cada contraseña 
df_weak["trivial_count"] = df_weak[trivial_cols].sum(axis=1)

#distribución (conteo)
count_dist = df_weak["trivial_count"].value_counts().sort_index()

print("Distribución de # de patrones triviales en contraseñas Fuertes:")
print(count_dist)

# gráfico
plt.figure(figsize=(8,5))
sns.barplot(x=count_dist.index, y=count_dist.values, palette="viridis")
plt.title("Número de patrones triviales en contraseñas Fuertes")
plt.xlabel("Cantidad de patrones triviales cumplidos")
plt.ylabel("Número de contraseñas")
plt.show()

# --- imprimir ejemplos por categoría ---
print("\nEjemplos de contraseñas débiles por número de patrones triviales:\n")
for n in sorted(df_weak["trivial_count"].unique()):
    subset = df_weak[df_weak["trivial_count"] == n]["pattern_template"].head(5).tolist()
    print(f"{n} patrones → {subset}")
```

```python
#filtrar solo contraseñas débiles 
df_strong = df[df["target_bin_zxcvbn"] == 0].copy()

#contar cuántos patrones triviales cumple cada contraseña 
df_strong["trivial_count"] = df_strong[trivial_cols].sum(axis=1)

#distribución (conteo)
count_dist = df_strong["trivial_count"].value_counts().sort_index()

print("Distribución de # de patrones triviales en contraseñas debiles:")
print(count_dist)

# gráfico
plt.figure(figsize=(8,5))
sns.barplot(x=count_dist.index, y=count_dist.values, palette="viridis")
plt.title("Número de patrones triviales en contraseñas debiles")
plt.xlabel("Cantidad de patrones triviales cumplidos")
plt.ylabel("Número de contraseñas")
plt.show()

# --- imprimir ejemplos por categoría ---
print("\nEjemplos de contraseñas fuertes por número de patrones triviales:\n")
for n in sorted(df_strong["trivial_count"].unique()):
    subset = df_strong[df_strong["trivial_count"] == n]["pattern_template"].head(5).tolist()
    print(f"{n} patrones → {subset}")

```

4. Entropia

```python
entropy_cols = ["entropy_shannon","entropy_nist",
                "entropy_renyi2","entropy_combinatorial","perplexity"]

print("Resumen estadístico de métricas de entropía:")
display(df[entropy_cols].describe().round(2))
```

```python
df_plot = df[["target_bin_zxcvbn"] + entropy_cols].melt(id_vars="target_bin_zxcvbn",
                                                 var_name="Métrica",
                                                 value_name="Valor")

plt.figure(figsize=(12,6))
sns.boxplot(data=df_plot, x="Métrica", y="Valor", hue="target_bin_zxcvbn")
plt.title("Comparación de métricas de entropía por grupo (débil=0, fuerte=1)")
plt.ylabel("Entropía / Perplejidad")
plt.xlabel("Métrica")
plt.legend(title="Grupo")
plt.show()
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(entropy_cols, 1):
    plt.subplot(2,3,i)
    sns.histplot(df[col], bins=50, kde=True, color="steelblue")
    plt.title(f"Distribución de {col}")
plt.tight_layout()
plt.show()
```

```python
# columnas de entropía ---
entropy_cols = ["entropy_shannon","entropy_nist",
                "entropy_renyi2","entropy_combinatorial","perplexity"]

# --- matriz de correlación ---
corr = df[entropy_cols].corr()

print("Matriz de correlación:")
print(corr.round(3))

# --- heatmap ---
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Mapa de correlación entre métricas de entropía")
plt.show()
```

```python
def category_pattern(row):
    cats = []
    if row["num_lower"] > 0: cats.append("lower")
    if row["num_upper"] > 0: cats.append("upper")
    if row["num_digits"] > 0: cats.append("digits")
    if row["num_symbols"] > 0: cats.append("symbols")
    return "+".join(cats)

df["cat_pattern"] = df.apply(category_pattern, axis=1)

top_patterns = df.groupby("target_bin_zxcvbn")["cat_pattern"].value_counts().groupby(level=0).head(5)
print(top_patterns)
```
